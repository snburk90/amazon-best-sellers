```python
from pyspark.sql import functions as F

# Total count of products per category
total_per_category = (
    df.groupBy("CATEGORIES")
      .agg(F.count("*").alias("total_products"))
)

# Count of products with rating > 4.48 per category
high_rating_per_category = (
    df.filter(F.col("RATING") > 4.48)
      .groupBy("CATEGORIES")
      .agg(F.count("*").alias("high_rated_products"))
)

# Join both counts on category
category_stats = (
    total_per_category.join(high_rating_per_category, on="CATEGORIES", how="left")
                      .fillna(0)  # Fill in 0 for categories that had no high-rated products
                      .orderBy(F.desc("high_rated_products"))
)

category_stats.display()
```

```python
from pyspark.sql.functions import col, round
category_stats = (
    category_stats.withColumn(
        "high_rating_ratio", 
        round(col("high_rated_products") / col("total_products"), 3)
    )
    .orderBy(col("high_rating_ratio").desc())
)

category_stats.display()
```

```python
top_brands = (
    df.filter(F.col("RATING") >= 4.4)
      .groupBy("BRAND")
      .count()
      .orderBy(F.desc("count"))
)

top_brands.display()
```

```python
profitable_df = df.filter(
    (F.col("RATING") >= 4.4) &
    (F.col("REVIEWS_COUNT") > 50) &  # good social proof
    (F.col("FINAL_PRICE") >= 15) &  # eliminate low-margin items
    (F.col("AVAILABILITY") == "In Stock")
)
profitable_df.display()
```

```python
from pyspark.sql.functions import col, log1p, round

scored_df = df.filter(
    (col("RATING") > 4.4) &
    (col("REVIEWS_COUNT").isNotNull()) &
    (col("FINAL_PRICE") > 0) &
    (col("NUMBER_OF_SELLERS").isNotNull()) &
    (col("FINAL_PRICE") > 15) &
    (col("NUMBER_OF_SELLERS") > 0) &
    (col("REVIEWS_COUNT") > 50) 
).withColumn(
    "opportunity_score",
    round((col("RATING") * log1p(col("REVIEWS_COUNT"))) /
    (col("FINAL_PRICE") * log1p(col("NUMBER_OF_SELLERS"))), 3)
)

scored_df.select("TITLE", "BRAND", "CATEGORIES", "RATING", "REVIEWS_COUNT", "FINAL_PRICE", "NUMBER_OF_SELLERS", "opportunity_score")\
    .orderBy(col("opportunity_score").desc())\
    .limit(10)\
    .display()
```
